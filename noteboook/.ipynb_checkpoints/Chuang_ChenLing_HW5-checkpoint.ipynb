#### (c) Convert the decision rules into a set of IF-THEN rules.
function <b>tree_to_code(tree, feature_names)</b> is referenced from https://www.kdnuggets.com/2017/05/simplifying-decision-tree-interpretation-decision-rules-python.html <br>
I have adjusted a bit so that it returns the absolute number of samples by multiplying by the number of samples reaching the root node, rather than showing the proportion of samples reaching a node for each class and for each output (tree_.value[])

```python
from sklearn.tree import _tree

def tree_to_code(tree, feature_names):
	"""
	Outputs a decision tree model as a Python function
	
	Parameters:
	-----------
	tree: decision tree model
		The decision tree to represent as a function
	feature_names: list
		The feature names of the dataset used for building the decision tree
	"""

	tree_ = tree.tree_
	feature_name = [
		feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
		for i in tree_.feature
	]
	print("def tree({}):".format(", ".join(feature_names)))

	def recurse(node, depth):
		indent = "  " * depth
		if tree_.feature[node] != _tree.TREE_UNDEFINED:
			name = feature_name[node]
			threshold = tree_.threshold[node]
			print("{}if {} <= {}:".format(indent, name, threshold))
			recurse(tree_.children_left[node], depth + 1)
			print("{}else:  # if {} > {}".format(indent, name, threshold))
			recurse(tree_.children_right[node], depth + 1)
		else:
			print("{}return {}".format(indent, tree_.value[node] * tree_.weighted_n_node_samples[node])) # adjust this line so that it returns the absolute number of samples

	recurse(0, 1)
```

```python
# IF-THEN rules for decision tree
tree_to_code(dt, list(df.columns))
```

```python
# IF-THEN rules for label powerset decision tree
tree_to_code(classifier_lp.classifier, list(df.columns))
```

#### (d) Use cost-complexity pruning to find a minimal decision tree and a set of decision rules with high interpretability.

```python
path = dt.cost_complexity_pruning_path(X, Y)
impurities = path['impurities']
alpha_list = path['ccp_alphas']
print(alpha_list)
acc_score = []

for i in alpha_list:
    tree_a = DecisionTreeClassifier(ccp_alpha=i)
    tree_a.fit(X,Y)
    y_pred = tree_a.predict(X)
    # acc_score.append(accuracy_score(y,y_pred))
    print('Alpha: ',i)
    print('Score',accuracy_score(Y,y_pred))
```

```python
best_alpha = alpha_list[1] # 0.04908333333333333

best_clf = tree.DecisionTreeClassifier(ccp_alpha= best_alpha)
best_clf.fit(X,Y)

dot_data = tree.export_graphviz(best_clf, out_file=None,feature_names=features,class_names=classes, filled=True, rounded=True,special_characters=True)  

#graph = graphviz.Source(dot_data)  
#graph
pydotplus.graph_from_dot_data(dot_data)

display(Image(data=graph.create_png()))
```

```python
# column 'ViolentCrimesPerPop' is our y
X_train = df_cc_train_imputed.drop(['ViolentCrimesPerPop'],axis=1)
Y_train = df_cc_train_imputed.ViolentCrimesPerPop
X_test = df_cc_test_imputed.drop(['ViolentCrimesPerPop'],axis=1)
Y_test = df_cc_test_imputed.ViolentCrimesPerPop

model_lr = LinearRegression()
model_lr.fit(X_train, Y_train)

Y_pred = model_lr.predict(X_test)

print(f"Test mean squared error: {mean_squared_error(Y_test,Y_pred)}")
```

#### (g) Fit a ridge regression model on the training set, with  chosen by cross-validation. Report the test error obtained.

```python
alphas = np.logspace(-6,6,100)  # Array of alpha values to try. Small positive values of alpha improve the conditioning of the problem and reduce the variance of the estimates
model_ridge = RidgeCV(alphas= alphas, cv = 5) 

model_ridge.fit(X_train, Y_train)
print(f"Best alpha: {model_ridge.alpha_}")

Y_pred_ridge = model_ridge.predict(X_test)

print(f"Test Mean Squared Error (Ridge): {mean_squared_error(Y_test, Y_pred_ridge)}")
```

#### (h) Fit a LASSO model on the training set, with  chosen by cross-validation. Report the test error obtained, along with a list of the variables selected by the model. Repeat with standardized features. Report the test error for both cases and compare them.

```python
model_lasso = LassoCV(alphas = alphas, cv = 5, max_iter=10000) # reuse alphas; max_iter to avoid ConvergenceWarning

model_lasso.fit(X_train, Y_train)
print(f"Best alpha (Lasso): {model_lasso.alpha_}")

Y_pred_lasso = model_lasso.predict(X_test)

print(f"Test Mean Squared Error (Lasso): {mean_squared_error(Y_test, Y_pred_lasso)}")

# getting the list of features
lasso_selected_features = []
for i in range(len(model_lasso.coef_)):
    if model_lasso.coef_[i] != 0:
        lasso_selected_features.append(X_train.columns[i])

print(f"Selected Variables: {lasso_selected_features}")
```

```python
# standardized features
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train) # fit_transform() method on training data 
X_test_std = scaler.transform(X_test) # transform() method on test data

model_lasso_std = LassoCV(alphas = alphas, cv = 5, max_iter=10000) # reuse alphas; max_iter to avoid ConvergenceWarning
model_lasso_std.fit(X_train_std, Y_train)

Y_pred_std = model_lasso_std.predict(X_test_std)

print(f"Test Mean Squared Error (Lasso Standardized Features): {mean_squared_error(Y_test, Y_pred_std)}")
```

#### Comparation
Given alpha range and cross-validation value, Lasso model has slightly better performance in terms of mean squared error than the standardized one. 

#### (i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
pca = PCA()
X_train_pca = pca.fit_transform(X_train_std)
```

```python
mse_list = list() # to record all mse
m = len(X_train.columns) # the number of principal components

# iterate through possible of m
for i in range(0, m):
    model = LinearRegression()
    k_fold_5 = model_selection.KFold(n_splits = 5, shuffle = True, random_state = 1) # 5-fold CV, with shuffle
    score = -1 * model_selection.cross_val_score(model, X_train_pca[:, 0:i+1], Y_train, cv = k_fold_5, scoring = 'neg_mean_squared_error').mean()
    mse_list.append(score)

# print(f"MSE list: {mse_list}")
min_mse = min(mse_list)
best_m = mse_list.index(min_mse)+1
print(f"The value of M selected by cross validation: {best_m} with the minimum MSE: {min_mse}")


best_pca = PCA(n_components=best_m)
X_train_bestM = best_pca.fit_transform(X_train)
X_test_bestM = best_pca.fit_transform(X_test)

model_bestM = LinearRegression()
model_bestM.fit(X_train_bestM, Y_train)
Y_pred_bestM = model_bestM.predict(X_test_bestM)

print(f"Test Mean Squared Error (PCR best M): {mean_squared_error(Y_test,Y_pred_bestM)}")
```

#### (j) In this section, we would like to fit a boosting tree to the data. As in classification trees, one can use any type of regression at each node to build a multivariate regression tree. Because the number of variables is large in this problem, one can use L1-penalized regression at each node. Such a tree is called L1 penalized gradient boosting tree. You can use XGBoost to fit the model tree. Determine  (the regularization term) using cross-validation.

```python
# Define the parameter grid for GridSearchCV
param_grid = {
    'alpha':  np.logspace(-6,6,13) # alpha range to test through
}

# XGBoost regressor
xgb_regressor = xgb.XGBRegressor(objective='reg:squarederror')

grid_search = model_selection.GridSearchCV(estimator = xgb_regressor, param_grid = param_grid, cv = 5, scoring = 'neg_mean_squared_error')
grid_search.fit(X_train, Y_train) # train XGBoost models

best_alpha = grid_search.best_params_['alpha']
print(f"Best alpha with XGBoost: {best_alpha}")

model_best_xgb = xgb.XGBRegressor(objective='reg:squarederror', alpha=best_alpha)
model_best_xgb.fit(X_train, Y_train)
Y_predbest_xgb = model_best_xgb.predict(X_test)

print(f"Test Mean Squared Error (XGB best alpha): {mean_squared_error(Y_test,Y_predbest_xgb)}")
```
